```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
evolutions = {}

class Pokemon:
    def __init__(self, num):
        url = "https://pokeapi.co/api/v2"
        data = requests.get(f"{url}/pokemon/{str(num)}").json()
        self.json = {}
        self.json['number'] = data['id']
        self.json['name'] = data['name']
        self.json['height'] = data['height']/10
        self.json['weight'] = data['weight']/10
        for t in data['types']:
            self.json['type_' + str(t['slot'])] = t['type']['name']
        if not data.get('type_2'):
            data['type_2'] = None
        for stat in data['stats']:
            self.json[stat['stat']['name']] = stat['base_stat']
        self.json['image_url'] = data['sprites']['front_default']
        species = requests.get(f"{url}/pokemon-species/{str(num)}").json()
        self.json['generation'] = species['generation']['name']
    
    def __getitem__(self, arg):
        return self.json[arg]
```

```python
columns = ['number', 'name', 'height', 'weight', 'type_1', 'type_2', 'hp', 'attack', 'defense', 'special-attack', 'special-defense', 'speed', 'generation', 'image_url']
data = []

for i in range(1, 808):
    pkm = Pokemon(i)
    print(pkm['number'], pkm['name'], end = '\r')
    data.append(pkm.json)

df = pd.DataFrame(data, columns=columns)
```

```python
df.to_csv('pokedex.csv', columns=columns, index=False)
```

```python
with open('pokedex.json', 'w+') as f:
    json.dump(data, f)
```
